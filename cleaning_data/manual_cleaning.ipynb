### Traitement des données
#### 1. Prix

```python
df['Prix'].unique()
```

```python
# Supprimer le symbole Euro (€), les espaces et les caractères de séparation de milliers
df['Prix_vente'] = df['Prix'].str.replace('€', '').str.replace('\u202f', '').str.replace(' ', '')

# Convertir la colonne 'Prix' en type de données float
df['Prix_vente'] = df['Prix_vente'].astype(float)

# Afficher les premières lignes du DataFrame mis à jour
df.head(3)
```

```python
df.dtypes
```

#### 2. Modèle

```python
df['Modèle'].unique()
```

```python
comptage_valeurs = df['Modèle'].value_counts()

print(comptage_valeurs)
```

```python
# Appliquer l'encodage one-hot sur la colonne 'Modèle'
df = pd.get_dummies(df, columns=['Modèle'], prefix='Modele')

# Supprimer la colonne originale 'Modèle'
#df.drop(columns='Modèle', inplace=True)

# Réinitialiser les index
df.reset_index(drop=True, inplace=True)
```

#### 3.Transmission

```python
df['Transmission'].unique()
```

On va garder pour la transmission
- boite / non boite
- boite manuelle / boite automatique

```python
TRANS_VALIDES = {'Boîte automatique', 'Boîte manuelle', '- Boîte', 'Boite non disponible'}
df.drop(df[~df["Transmission"].isin(TRANS_VALIDES)].index, inplace=True)
df.reset_index(inplace=True, drop=True)
TYPE_BOITE = {'Boîte automatique', 'Boîte manuelle'}
df['Boite'] = df['Transmission'].apply(lambda x: 1 if x in {'Boîte automatique', 'Boîte manuelle'} else 0)
#data['type_boite'] = df['Transmission'].map({'Boîte automatique': 1, 'Boîte manuelle': 0})
df['type_boite']= df["Transmission"].isin(TYPE_BOITE)
#df.drop(columns='Transmission', inplace= True)
df.reset_index(inplace= True, drop=True)
```

```python
df.head(3)
```

#### 4. Kilometrage

```python
df['Kilométrage'].unique()
```

```python
df['Kilométrage (km)'] = df['Kilométrage'].str.replace(' km', '').str.replace('\u202f', '').str.replace(' ', '').astype(int)
#df.drop(columns='Kilométrage', inplace= True)
#df.reset_index(inplace= True, drop=True)
```

```python
df.head(3)
```

#### 5. Carburant

```python
df['Carburant'].unique()
```

```python
valeur_carb = df['Carburant'].value_counts()
valeur_carb
```

array(['Essence', 'Electrique/Essence', 'Diesel', 'Autres',
       'Electrique/Diesel', 'Electrique', '- Carburant', 'CNG'],
      dtype=object)

On va garder
- essence
- diesel
- electrique ou/et  non (hybride)

On va donc recoder les modalitées de `Carburant`

```python
df['Carburant'] = df['Carburant'].replace({'- Carburant': 'Autres', 'CNG': 'Autres'})
df['Carburant'] = df['Carburant'].replace({'Electrique/Essence': 'Hybride', 'Electrique/Diesel': 'Hybride'})
df = pd.get_dummies(df, columns=['Carburant'], prefix='Carburant')
```

```python
df.head(2)
```

#### 6. Puissance

```python
df['Puissance (KW)'] = df['Puissance'].str.extract('(\d+)', expand=False)
df['Puissance (KW)'] = pd.to_numeric(df['Puissance (KW)'], errors='coerce')
#df.drop(columns=['Puissance'], inplace=True)
```

```python
df.head(2)
```

```python
df.drop(columns=["Prix", "Transmission","Kilométrage", "Puissance","Version", "Vendeur", "Nom de la Voiture", "Évaluations", "Date"], inplace=True)
df.reset_index(drop=True, inplace=True)
```
